```python
#딕셔너리를 Seires data로 만들기
dic = {'chicken': 20000, 'pizza': 25000, 'rice': 1000, 'coke': 2000}
srdic = pd.Series(dic)
srdic
```

```python
#srdic의 이름 설정, 인덱스 이름 설정
srdic.name = 'Food'
srdic.index.name = 'Kind'
srdic
```

```python
#srdic의 인덱스 변경
srdic.index = ['치킨', '피자', '밥', '콜라']
srdic
```
