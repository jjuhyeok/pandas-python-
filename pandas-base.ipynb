# 6 : NaN교체

```python
#행의 값중 하나라도 nan인 경우 그 행을 없애는 코드
df.dropna(how='any')
```

```python
# 행의 값의 모든 값이 nan인 경우 해당하는 행 삭제
df.dropna(how='all')
```

drop 함수는 기존의 DataFrame을 변화시키지 않는다
만약 기존의 DataFrame을 변화시키려면
inplace=True라는 인자를 추가하면 DataFrame도 바뀌게 된다

```python
# nan값에 값 넣기
df.fillna(value=99)
```

```python
# nan확인하는 코드
# nan이면 True로 표시됨
df.isnull()
```

```python
# F열에서 nan값을 포함하는 행만 추출하기
df.loc[df.isnull()['F'],:]
```

```python
pd.to_datetime('20160701')
```

```python
# 특정 행 drop하기
#df.drop(pd.to_datetime('20160702'))
df.drop(['2016-07-02']) #같은 의미
```

```python
# 2개 이상도 가능
df.drop([pd.to_datetime('20160701'),pd.to_datetime('20160705')])
```

```python
# F 열 삭제하기
df.drop('F', axis =1)
```

```python
# 2개 이상의 열도 가능
df.drop(['B','D'], axis = 1)
```

# 7 : 데이터 분석용 함수

```python
data = [[27, np.nan],
           [71, -4.2],
        [np.nan, np.nan],
        [1.25, -7.8]]
df = pd.DataFrame(data, columns=["one", "two"], index=["a", "b", "c", "d"])
df
```

```python
# 합 함수
df.sum(axis = 0) #행 방향으로의 합(한 열의 합)
```

```python
#열 방향으로의 합
df.sum(axis = 1)
```

NaN의 값은 제외시키고 계산한다

```python
#NaN의 값을 포함시키려면 skipna 함수를 사용하면 된다
df.sum(axis=1, skipna=False)
```

```python
#특정 행이나 열에서만 계산하기
df['one'].sum()
```

```python
df.loc['b'].sum()
```
