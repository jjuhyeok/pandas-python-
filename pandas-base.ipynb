# 2 : DataFrame

```python
data = {'name': ['Beomwoo', 'Beomwoo', 'Beomwoo', 'Kim', 'Park'],
        'year': [2013, 2014, 2015, 2016, 2015],
        'points': [1.5, 1.7, 3.6, 2.4, 2.9]}
df = pd.DataFrame(data)
df
```

행과 열의 구조를 가진 데이터

```python
#행 방향의 index
df.index
```

```python
#열 방향의 index
df.columns
```

```python
#값 얻기
df.values
```

```python
#각 인덱스에 대한 이름 설정
df.index.name = 'Num'
df.columns.name = 'Info'
df
```

```python
#DataFrame을 만들면서 행과 열을 설정하는 법
df2 = pd.DataFrame(data, columns=['year', 'name', 'points', 'penalty'],
                  index=['one', 'two', 'three', 'four', 'five'])
df2
```

penalty는 data에 정의되어있지 않기 때문에 NaN값이 뜬다

```python
#계산 가능한 값들에 대한 설명
df2.describe()
```

```python
#indexing
data = {"names": ["a", "b", "c", "d", "e"],
           "year": [1999, 2000, 2001, 2002, 2003],
           "points": [90, 80, 30, 70, 99]}
df = pd.DataFrame(data, columns=["year", "names", "points", "penalty"],
                          index=["one", "two", "three", "four", "five"])
df
```

```python
#year 열을 선택
df['year']
```

```python
#df['year']와 같은 역할을 하는 코드
df.year
```

```python
df[['year','points']]
```

```python
#df['penalty'] = 0.5 밑에랑 똑같은 코드
df.penalty = 0.5
df
```

```python
#새로운 열을 추가하기
#np.arange(시작점,끝점) 0,5로 하면 0~4까지  밑에처럼 그냥 5로하면 0~4까지
#df['zeros'] = np.arange(5)
df.zeros = np.arange(5) 
df
```

```python
# Series 추가
val = pd.Series([-1.2, -1.5, -1.7], index=['two','four','five'])
df['debt'] = val
df
```

```python
#열 삭제하기
del df['names']
df
```

```python
#인덱스 이름 추가하기
df.index.name = 'Order'
df
```

```python
#열 이름 추가하기
df.columns.name = 'Info'
df
```
